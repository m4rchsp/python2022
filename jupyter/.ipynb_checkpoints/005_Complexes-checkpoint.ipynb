# Complexes
complex, <class 'complex'>

Ce sont les nombres imaginaires de la forme
$$a + b j$$
Où $j^2=-1$ et a,b sont réels.
## Littéraux

```python
2+4j, (2+4j)**2, 3*(2+4j)
```

```python
type(3j), type(3*j)
```
